```python
%pylab inline
#load the list of points
obs_data = loadtxt("movimiento.dat")
x_obs = obs_data[:,0]
y_obs = obs_data[:,1]
```

```python
scatter(x_obs,y_obs)
```

```python
def likelihood(y_obs, y_model):
    chi_squared = sum((y_obs-y_model)**2) #And what about the variance?? We don't take it into account
    return chi_squared
```

```python
def my_model(x_obs, a, b, c):
    
    return ((a*(x_obs)**2) + b*(x_obs) + c)
```

```python
a_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))
c_walk = empty((0))

a_walk = append(a_walk, random.random())
b_walk = append(b_walk, random.random())
c_walk = append(c_walk, random.random())

print a_walk
print b_walk
print c_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make

for i in range(n_iterations):
    a_prime = np.random.normal(a_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)
    c_prime = np.random.normal(c_walk[i], 0.1)

    y_init = my_model(x_obs, a_walk[i], b_walk[i], c_walk[i])
    y_prime = my_model(x_obs, a_prime, b_prime, c_prime)

    alpha = likelihood(y_obs, y_prime) - likelihood(y_obs, y_init)
    
    if(alpha<=0.0):
        a_walk  = append(a_walk,a_prime)
        b_walk  = append(b_walk,b_prime)
        c_walk  = append(c_walk,c_prime)
    
    else:
        beta = random.random() #Ya me lo define entre 0 y 1
        if(beta<=exp(alpha)):
            a_walk = append(a_walk,a_prime)
            b_walk = append(b_walk,b_prime)
            c_walk = append(c_walk,c_prime)
        else:
            a_walk = append(a_walk,a_walk[i])
            b_walk = append(b_walk,b_walk[i])
            c_walk = append(c_walk,c_walk[i])
```
